```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
flowers = load_iris()
```

```python
dir(flowers)
```

```python
flowers.target_names
```

```python
df = pd.DataFrame(flowers.data, columns=flowers.feature_names)
df['target'] = flowers.target
df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=100_000)
```

```python
model.fit(flowers.data, flowers.target)
```

```python
model.score(flowers.data, flowers.target)
```

```python
model.predict([[10,10,10,10]])
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier().fit(flowers.data, flowers.target)
```

```python
tree.score(flowers.data, flowers.target)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=[15,15])
plot_tree(tree, feature_names=flowers.feature_names, class_names=flowers.target_names, filled=True, )
plt.show()
```
