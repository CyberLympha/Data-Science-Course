#### countplot

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.show()
```

#### catplot

```python
sns.catplot(x='Survived', y='Age', col='Pclass',
               data=df, kind="violin",
               col_wrap=4, height=3, aspect=.8);

sns.catplot(x='Survived', y='Age', col='Pclass',
               data=df, kind="box",
               col_wrap=4, height=3, aspect=.8);
```

#### lmplot

```python
sns.lmplot(x='Age', y='Fare', data=df, hue='Survived', fit_reg=False)
plt.show()
```

#### pairplot

```python
sns.pairplot(data=df, kind='scatter')
plt.show()
```

#### jointplot

```python
sns.jointplot(x=df.Pclass, y=df.Fare, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='Age', y='Fare', data=df, kind='kde', color='g')
plt.show()
```
